I will be creating a graph using NetworkX package from Python

```python
import networkx as nx
import matplotlib.pyplot as plt
```

I will create an empty graph and add the nodes to it first

```python
G = nx.Graph()
nodes = ['Andre', 'Beverly', 'Carol', 'Diane', 
        'Ed', 'Fernando', 'Heather', 'Ike', 'Jane']
G.add_nodes_from(nodes)
```

I will then create edges of the Graph which basically represents the connection between the nodes.
After createing the tuples which represents the edges , I will add them to the Graph which was created in the earlier step

```python
edges = [('Andre', 'Beverly'),
         ('Andre', 'Carol'),
         ('Andre', 'Diane'),
         ('Andre', 'Fernando'),
         ('Beverly', 'Diane'),
         ('Beverly', 'Garth'),
         ('Beverly', 'Ed'),
         ('Carol', 'Fernando'),
         ('Carol', 'Diane'),
         ('Ed', 'Diane'),
         ('Ed', 'Garth'),
         ('Diane', 'Fernando'),
         ('Diane', 'Garth'),
         ('Fernando','Garth'),
         ('Heather', 'Fernando'),
         ('Heather', 'Garth'),
         ('Heather', 'Ike'),
         ('Ike','Jane' )
        ]
G.add_edges_from(edges)
```

Finally draw the Graph

```python
nx.draw(G,with_labels=True)
plt.show()
```
